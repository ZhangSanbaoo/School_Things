# Exercise 7

- Compute the number of diamonds for each cut group in the dataframe *df_diamonds* using the price column.

- Compute the minimum, the maximum and the mean price for each cut group of the dataframe *df_diamonds* 

- Compute the minmum, the maximum and count of price and table for each cut group of the dataframe *df_diamonds* . Use `agg()`.

```python
cut_groups = df_diamonds.groupby('cut')
count_by_cut = cut_groups['price'].count()
print(count_by_cut)
print('\n\n')
cut_groups = df_diamonds.groupby('cut')
price_stats = cut_groups['price'].agg(['min', 'max', 'mean'])
print(price_stats)
print('\n\n')
cut_groups = df_diamonds.groupby('cut')
price_table_stats = cut_groups.agg({'price': ['min', 'max', 'count'], 'table': ['min', 'max', 'count']})
price_table_stats
```

## Exercice 8
- If you want to buy one diamond, wich cut should you buy ? Look for the best quality/Price. Explain what you did.

```python
cut_groups = df_diamonds.groupby('cut') #We use groupby function to group elements by cut.
avg_price_per_carat = cut_groups['price'].mean() / cut_groups['carat'].mean() #We calculate the price by carat.
avg_price_per_carat = avg_price_per_carat.sort_values() #We sort the result by order.
print(avg_price_per_carat) #We can see that the Fair quality has the least price per carat, so maybe Fair cut.
```

## Exercice 10
- Use a crosstabulation on columns cut & price of the dataframe *df_diamonds*

- Read documentation and play with it.

```python
cross = pd.crosstab(df_diamonds['cut'], df_diamonds['price'])
print('**********\n',cross)
#Use margins option to create a new column which shows the sum number of diamond for that cut.
cross = pd.crosstab(df_diamonds['cut'], df_diamonds['price'], margins=True)
print('\n**********\n',cross)
#Use normalize option to show frequency instead of number of occurrences.
cross = pd.crosstab(df_diamonds['cut'], df_diamonds['price'], normalize='columns')
print('\n**********\n',cross)
#Create a table where the values represent the average price of diamonds with a particular cut and price value.
#With the 'aggfunc' option to specify a function to apply to the values in the table.
cross = pd.crosstab(df_diamonds['cut'], df_diamonds['price'], values=df_diamonds['price'], aggfunc='mean')
print('\n**********\n',cross)
```
